```python
import pandas as pd
import pymongo
import matplotlib.pyplot as plt
from datetime import datetime
from pymongo.errors import PyMongoError

# 1. Conexión a MongoDB Atlas
def conectar_mongo():
    try:
        client = pymongo.MongoClient(
            "mongodb+srv://antcmara04:Matthew_1908**@infinitteattack.luqf1uc.mongodb.net/test"
            "?retryWrites=true&w=majority&appName=InfinitteAttack",
            connectTimeoutMS=5000,
            serverSelectionTimeoutMS=5000
        )
        # Verificar conexión
        client.admin.command('ping')
        print("✅ Conexión exitosa a MongoDB Atlas")
        return client.test.partidas  # Base de datos: test, Colección: partidas
    except PyMongoError as e:
        print(f"🔴 Error de conexión: {str(e)}")
        return None

# 2. Obtener y preparar datos
def obtener_datos():
    try:
        collection = conectar_mongo()
        if collection is None:
            return None
            
        # Consulta todos los documentos en la colección partidas
        datos = list(collection.find({}))
        
        if not datos:
            print("⚠️ No se encontraron documentos en la colección 'partidas'")
            return None
            
        # Convertir a DataFrame
        df = pd.DataFrame(datos)
        
        # Procesamiento de fechas (para el formato MongoDB)
        if 'fecha' in df.columns:
            df['fecha'] = pd.to_datetime(df['fecha'].apply(lambda x: x['$date'] if isinstance(x, dict) else x))
        
        return df
    
    except Exception as e:
        print(f"🔴 Error al procesar datos: {str(e)}")
        return None

# 3. Generar gráfico lineal interactivo
def generar_grafico():
    df = obtener_datos()
    if df is None or df.empty:
        print("No hay datos para mostrar")
        return
    
    # Configurar el gráfico
    plt.figure(figsize=(12, 6))
    
    # Gráfico lineal
    plt.plot(df['fecha'], df['objetosRecogidos'], 
             marker='o', 
             linestyle='-',
             color='#3498db',
             linewidth=2,
             markersize=8,
             label='Objetos Recogidos')
    
    # Personalización
    plt.title('Historial de Objetos Recogidos', pad=20, fontsize=16)
    plt.xlabel('Fecha', fontsize=12, labelpad=10)
    plt.ylabel('Cantidad', fontsize=12, labelpad=10)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.legend()
    
    # Formato de fechas
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Mostrar gráfico
    plt.show()

# Ejecución principal
if __name__ == '__main__':
    generar_grafico()
```
